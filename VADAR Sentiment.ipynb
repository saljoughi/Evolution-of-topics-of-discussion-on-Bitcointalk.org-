```python

import re
import numpy as np
import pandas as pd

import csv


# spacy for lemmatization
import spacy



import warnings
warnings.filterwarnings("ignore",category=DeprecationWarning)





from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyser = SentimentIntensityAnalyzer()

def sentiment_analyzer_scores(sentence):
    score = analyser.polarity_scores(sentence)
    #print("{:-<40} {}".format(sentence, str(score)))
    return(score['compound'])
   
```

```python
def Sentiment_code (serie): 
    timeframe=serie
    inputpath='/home/davoud/Documents/cs project/comment source/edited data oct/'
    input_here= inputpath+timeframe+'.csv'
    pathldavis = '/home/davoud/Documents/cs project/trial/Nov29' + timeframe + '.html'
    pathshortsumary2= '/home/davoud/Documents/cs project/trial/Nov29/' + timeframe  +'_Sentiment_table'+  '.csv'
    import re
    import numpy as np
    import pandas as pd

    import csv
    #feeding  data source

    df = pd.read_csv(input_here)
    # step 7 Data Cleaning
    # Convert to list
    data = df.for_lda.values.tolist()

    #print ('Initial Number of Comments = ',len(data) )

    # Remove Emails
    # Remove new line characters
    # Remove distracting single quotes
    k=0
    temp100=[]
    temp_comments=[]
    temp_cleaned=[]
    for sent in data:
        try: 
            sent=str(sent)
            #sent_comments=sent
            senttiments_compound_score = sentiment_analyzer_scores(sent)
            sent= sent.lower()
            sent = re.sub(r'http\S+', ' ', sent) # Remove websites
            sent=  re.sub(r'\S+.(com|org)\S+', '', sent) # Remove website2
            
            sent = re.sub('\S*@\S*\s?', '', sent) # Remove Emails
            sent = re.sub('\s+', ' ', sent) # Remove new line characters
            sent = re.sub("\'", "", sent) # Remove distracting single quotes
            sent= re.sub( "[^a-zA-Z]", " ", sent)
            if len(sent.split())>5 :
                temp100.append(senttiments_compound_score)
                #temp_comments.append(sent_comments)
                temp_cleaned.append(sent)
                k +=1

        except TypeError:
            continue 

    import pandas as pd   
    
        
    
        
    # dictionary of lists   
    #dict = {'name': nme, 'degree': deg, 'score': scr}   
    dict = {'Compound_Score':temp100, 'Edited_Post':temp_cleaned}     
    df123 = pd.DataFrame(dict)  
        
    # saving the dataframe  
    df123.to_csv(pathshortsumary2) 
```
